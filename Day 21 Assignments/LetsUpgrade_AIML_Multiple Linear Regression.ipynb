```python
a = dataset.iloc[:, :1]
b = dataset.iloc[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
b_train, b_test, a_train, a_test = train_test_split(b ,a, test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
mul_reg = LinearRegression()
```

```python
mul_reg.fit(b_train, a_train)
```

```python
mul_reg.coef_
```

```python
mul_reg.intercept_
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
apred = mul_reg.predict(b_test)
```

```python
RMSE = np.sqrt(mean_squared_error(a_test, apred))
r_square = r2_score(a_test, apred)

print('The R^0.5 value is:', r_square)
print('The RMSE value is:', RMSE)
```
